```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\query prossesing\alphabet_stock_data.csv")

print(df)

df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

data = df[(df["Date"] >= "2020-04-01") & (df["Date"] <= "2020-04-30")]

plt.figure(figsize=(10,5))
plt.plot(data["Date"], data["Close"], marker='o')
plt.title("Historical Stock Prices of Alphabet Inc.")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\query prossesing\alphabet_stock_data.csv")

print(df)


df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)


data = df[(df["Date"] >= "2020-04-01") & (df["Date"] <= "2020-04-30")]

plt.figure(figsize=(10,5))
plt.bar(data["Date"].dt.strftime("%d-%m-%Y"), data["Volume"])


plt.title("Trading Volume of Alphabet Inc.")
plt.xlabel("Date")
plt.ylabel("Volume")

plt.xticks(rotation=45)

plt.show()
```
